#Avoid adding items one by one inside a loop. Each time you "append" to a Series, Pandas creates a full copy of the data in memory.

```python
s2 = pd.Series([], dtype='float64')
s2.at['a'] = 0
print(s2)
s3=pd.Series([1,2,3,4],index=['a','b','c','d'])
print(s3)
s4=pd.concat([s2,s3])
print(s4)
a=pd.Series(['p','q','r','s'],index=[11,12,13,14])
print(a)
```

#**Data Frames**

```python
import numpy as np
```

```python
data = np.random.randint(1, 100, size=(10, 10))

# Naming columns C1, C2... C10 and rows R1, R2... R10
dataf = pd.DataFrame(data,
                  columns=[f'C{i}' for i in range(1, 11)],
                  index=[f'R{i}' for i in range(1, 11)])

print(dataf)
```

#Applying operations on above df

```python
print(dataf[dataf['C4']>50])
print(dataf.mean())      # Mean of each column
print(dataf.sum(axis=1)) # Sum of each row
print(dataf.iloc[0:3, 0:5]) #0 to 3 column and 0 to 5 rows
dataf['C2']=dataf['C2']+100
dataf['C1'].rename('First')
print("\nColumn 2: ",dataf.iloc[:, 1:2])
print("\nTop n largest values: ",dataf.nlargest(2,'C3'))
print("\nUnique values: ",dataf['C2'].unique())
print("\nUnique values: ",dataf['C3'].value_counts(normalize=True)) #show percentage values instead of just count

```

using data set without uploading through url

```python
import pandas as pd
url = "https://raw.githubusercontent.com/NotAyushXD/Titanic-dataset/refs/heads/master/train.csv"

df = pd.read_csv(url)
print(df.head())

```

#**analyze dataset**

```python
print(df.head())  #first five rows
print()
print(df.tail())  #last five rows
print()
print(df.describe()) #desciption of data
print()
print(df.shape)  #dimensions of data
print()
print(df.info())
print()
print("null entries in the dataset: ",df.isnull().sum())
print()
print(df.columns)
```

#showing the data of survivors

```python
survivors=df[df['Survived']==1]
print(survivors.head())
print()
print(survivors.value_counts())
```

#various data analysis functions

```python
print(df['Fare'].mean())
print()
print(df['Sex'].value_counts())
print()
print("Average age of passegers:",df['Age'].mean())
print("\nSurvival Rate by Gender:")
print(df.groupby('Sex')['Survived'].mean())
print()
print(df.groupby('Sex')['Survived'].value_counts())
```

#Handling Missing Values

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df.drop(columns=['Cabin'], inplace=True)  # drop cabin column (too many null values)
```
