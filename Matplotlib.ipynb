```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
```

#Matplotlib basics

```python
plt.style.use('dark_background')
x=np.arange(-10,11)
y=x**2

plt.figure(figsize=(6,6))
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x^2')
plt.plot(x,y)
plt.show()
```

#Basic Plotting Functions
'ro--': Red color, O marker, Dashed line.
'g^:': Green color, Triangle marker, Dotted line.
'ks-': B(k)lack color, Square marker, Solid line.

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(x)
plt.figure(figsize=(12,6))
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=sin(x)')
plt.plot(x,y,"ro--")  #Creates a line chart. You can choose color by adding attribute color
plt.show()
```

line chart for comparison

```python
print(plt.style.available) # Lists all themes
data = np.random.randint(1, 100, size=(10, 10))
df = pd.DataFrame(data,
                  columns=[f'C{i}' for i in range(1, 11)],
                  index=[f'R{i}' for i in range(1, 11)])
plt.figure(figsize=(8, 4))
plt.plot(df['C1'], label='Column 1', marker='o')
plt.plot(df['C2'], label='Column 2', linestyle='--')
plt.plot(x, y)
plt.annotate('Peak', xy=(1.57, 1), xytext=(3, 1.2),
             arrowprops=dict(facecolor='red', shrink=0.05)) #to write text on an arrow
#plt.xlim(min, max) / plt.ylim(min, max): Manually set the boundaries of your graph..........................

#plt.xticks() / plt.yticks(): Manually set what numbers or labels appear on the axes.......................

plt.title("Comparison of C1 and C2")
plt.xlabel("Row Index")
plt.ylabel("Values")
plt.legend()
plt.show()
```

scatter plot

```python
plt.figure(figsize=(8,4))
plt.scatter(df.index, df['C1'], label='Column 1', color='blue', marker='o')
plt.scatter(df.index, df['C2'], label='Column 2', color='red', marker='x')

plt.title("Scatter Plot: C1 vs C2")
plt.xlabel("Row Index")
plt.ylabel("Values")
plt.legend()  #Adds a box identifying different data series (works if you used the label parameter in your plot)
plt.show()
```

#Subplot

```python
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(x, y)

plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.scatter(df.index, df['C1'])

plt.tight_layout()   # Prevents overlapping labels
plt.figure(figsize=(8,6))
plt.show()
```

#**plt.boxplot():**
Shows the median and outliers (crucial for statistics).
#**plt.stackplot():**
Stacked areas (good for showing how parts of a total change over time).
#**plt.errorbar():**
Adds "whiskers" to your points to show uncertainty.


```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, patch_artist=True) # patch_artist fills the boxes with color
plt.title('Box Plot of 3 Groups')
plt.show()
print()
# X-axis (Time)
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

# Y-axis data (Each list is a "layer" in the stack)
dept_tech = [40, 55, 45, 60]
dept_sales = [30, 25, 35, 30]
dept_hr = [10, 15, 10, 10]

plt.figure(figsize=(10, 6))

# Define colors for each layer
colors = ['skyblue', 'lightgreen', 'salmon']

# Create the stackplot
plt.stackplot(quarters, dept_tech, dept_sales, dept_hr,
              labels=['Tech', 'Sales', 'HR'],
              colors=colors, alpha=0.8)

plt.title('Profit Contribution by Department')
plt.xlabel('Quarters')
plt.ylabel('Profit (in Thousands)')
plt.legend(loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.5)
```

#bar chart ,histogram ,pie chart
Attribute function- width, edgecolor, linestyle, linewidth.

```python
subject=['hindi','english','maths','science']
marks=[90,80,70,60]
colors=['red','blue','green','yellow']
plt.bar(subject,marks,color=colors)
plt.show()

print()

points=np.random.randint(1,100,50)
bins=np.arange(0,101,5)
plt.hist(points, bins=bins, edgecolor='red', color='skyblue')

plt.show()

print()

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [45, 25, 15, 15]  # These should ideally add up to 100
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.figure(figsize=(8, 8))

# Create the pie chart
plt.pie(sizes,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%', # Shows the percentage on the slice
        startangle=140,    # Rotates the start of the chart
        shadow=True,
        wedgeprops={'edgecolor': 'black', 'linewidth': 2})       # Adds a 3D shadow effect

plt.title("Programming Language Popularity")
plt.show()
```
