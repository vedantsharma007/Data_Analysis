need to import the module

```python
import numpy as np
```

```python
arr=np.array([[1,2,8],[4,7,6]])
print(arr)
print(type(arr))
```

all funtions of np array

```python
print(arr.size)  #no.of elements in array
print(arr.shape) #this will print (no.of rows,no.of columns)
print(arr.dtype) #data type
print(arr.transpose())
print(arr.ndim) #no.of dimensions
print(arr.max)
print(arr.min)
print(arr.argmax) # returns index of maximum no.
print(np.sum(arr)) #sum of all elements
print(np.sum(arr,axis=1)) #sum of each row axis=0 means sum of each column
```

more methods in array


```python
print(arr.mean())   # np.method()==Array_name.Method_name()
print(np.sort(arr))
print(np.std(arr))
print(np.log(arr))
print(np.sqrt(arr))
```

```python
nums=np.ones(5) #similarly for np.zeroes and np.empty(2,4) will create empty array
print(nums)
nums2=np.arange(1,11)  #third parameter for skip index
print(nums2)
print(arr.reshape((3,2)))
print(arr.flatten()) #ravel is used for only reference  flatten returns copy of original
```

Mathematical operations on two matrices/arrays

```python
a1=np.array([[1,2,3],[4,5,6]])
a2=np.array([[2,5,2],[1,6,5]])
a3=np.array([[2,2],[1,6],[3,4]])
print(a1+a2)  # np.add(a1,a2)
print(a1-a2)  #np.subtract(a2,a1)
print(a2/a1)
print(a1*a2) # simple multiplication
print(a1@a3) #matrix multiplication
print(a1.dot(a3)) # same as matrix multiplication
```

Random

```python
print(np.random.random(1)) #no.of element is 1
print(np.random.random((2,2)))  # 2*2 matrix
print(np.random.randint(1,11,(3,3))) #range 1 to 10 and 3*3 matrix of integers
print(np.random.rand(2,2)) # float values
print(np.random.randn(2,2)) # negative numbers
print(np.random.choice(nums2)) #choose one element from 1d array
```
