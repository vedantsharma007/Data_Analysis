```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
```

#Seaborn Basics

```python
roll_no=np.arange(1,11)
marks=np.random.randint(1,101,(10))
marks1=np.random.randint(1,101,(10))
df=pd.DataFrame({'roll_no':roll_no,'marks':marks,'marks1':marks1})
print(df)
```

#sns line plot


```python
sns.lineplot(x='roll_no',y='marks',data=df,label='Student1',marker='o',color='blue')
plt.title('Student marks')
sns.lineplot(x='roll_no',y='marks1',data=df,label='Student2',marker='*',color='green')
```

#scatter plot

```python
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
#hue is the parameter used for Color Encoding. It tells Seaborn to "split" your
#data based on a specific column and give each group a different colour.

plt.show()
```

#**Key Features of displot:**
#**kde**
True: Adds a Kernel Density Estimate line (the smooth curve) over the histogram. This helps you see the "shape" of the data.
#**bins**
Controls the number of bars.
#**kind**
You can switch from a histogram to a density curve only by setting kind="kde".
#**hue**
If you combine your marks into one column, you can use hue to compare the distribution of Subject 1 vs Subject 2 in different colors.

```python
sns.displot(df['marks'], bins=5, color='purple', kde=True)
plt.title("Distribution of Subject 1 Marks")
plt.show()
```

#Comparing Two Distributions

```python
sns.kdeplot(df['marks'], fill=True, label='Student 1')
sns.kdeplot(df['marks1'], fill=True, label='Student 2')

plt.title("Score Density Comparison")
plt.legend()
plt.show()
```

Histogram: Shows raw counts in "buckets."

KDE (Curve): Shows the probability density (useful for seeing if data is "Normal" or "Skewed").

#**Heatmaps**
annot=True: Displays the actual passenger count in each cell.

fmt="d": Ensures the numbers are printed as integers (decimals can make it look messy).

cmap="YlGnBu": Uses a "Yellow-Green-Blue" color scale, which is standard for showing intensity (peaks)

```python
flight_df = sns.load_dataset('flights')
flight_pivot = flight_df.pivot(index="month", columns="year", values="passengers")

#  Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flight_pivot, annot=True, fmt="d", cmap="YlGnBu")

plt.title("Monthly Air Travel Passengers (1949-1960)")
plt.show()
```

#Overlaying / combining line plot and scatter plot together

```python
plt.figure(figsize=(10, 5))

sns.lineplot(data=df, x='roll_no', y='marks', color='blue', label='Trend Line')

sns.scatterplot(data=df, x='roll_no', y='marks', color='red', s=100, label='Actual Marks')

plt.title("Combined Line and Scatter Plot")
plt.show()
```
